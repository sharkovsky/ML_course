# Generate 2D data set

```python
N = 10  #number of points
D = 2   #dimensionality
data = np.random.rand(N,D)
plt.xlim(0, 1)   #xmin-xmax
plt.ylim(0, 1)   #ymin-ymax
plt.scatter(data[:,0],data[:,1])
print(np.mean(data[:,0]))
```

# TaskA - Matrix Standardization

```python
# standardize the data, i.e. subtract the mean and divide by the standard deviation for each dimension

newdata = np.zeros(shape=(N,D))

for d in range(0,D) :  #iterate over dimensions
    mean   = np.mean(data[:,d])
    stddev = np.std (data[:,d])
    print('before: dim_', d, ': mean ' , mean , ', stddev ' , stddev, ';')
    newdata[:,d] = data[:,d] - mean
    newdata[:,d] = data[:,d] / stddev
    newmean   = np.mean(newdata[:,d])
    newstddev = np.std (newdata[:,d])
    print('after : dim_', d, ': mean ' , newmean , ', stddev ' , newstddev, ';')

plt.scatter(newdata[:,0],newdata[:,1])
```
